```python
x = tips['total_bill']
y = tips['tip']
```

```python
sns.jointplot(x = x, y=y)
```

```python
sns.set()
sns.jointplot(x = x, y=y, kind = 'hex')
```

```python
sns.jointplot(x = x, y = y, kind = 'kde')
```

```python
f, ax = plt.subplots(figsize = (6,6))
cmap = sns.cubehelix_palette(as_cmap = True, dark = 0, light = 1, reverse= True)
sns.kdeplot(x, y, cmap = cmap, n_levels=60, shade=True)
```

```python
g = sns.jointplot(x, y, kind = 'kde', color = 'r')
g.plot_joint(plt.scatter, c = 'w', s = 30, linewidth = 1, marker = '+')
g.ax_joint.collections[0].set_alpha(0)
```

```python
g = sns.jointplot(x, y, kind = 'kde', color = 'r')
g.plot_joint(plt.scatter, c = 'w', s = 30, linewidth = 7, marker = '+')
g.ax_joint.collections[0].set_alpha(0)
```

```python
## keep data in x -axis and parameters or calculations in y-axis
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x+i*0.5)*(7-i))
```

```python
sinplot()
```

```python
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(2):
        plt.plot(x, np.sin(x+i*0.5)*(7-i))
```

```python
sinplot()
```

```python
def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x))
```

```python
sinplot()
```
